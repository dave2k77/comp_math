# Nonlinear Functions and their Transformations II

**Introduction**

As in previous examples, functions may undergo stretches or compressions. Let $A > 0$ and $f(x)$ be a function of $x$. Let us define $F(x)$ as $F(x) = A \cdot f(x)$. Then $F(x)$ is a vertical stretch of the function $f(x)$ by a scale factor of $A$. If $0 < A < 1$, then $F(x)$ is a vertical compression of the function $f(x)$. 

**Example**

Sketch the graph of the function $f(x) = 3 + \frac{2}{x^2}$, $x \ne 0$.

Let us first define the functions $f(x)$, $g(x)$ and $h(x)$ as follows:

```python
def f(x):
    if (x == 0):
        return None
    else:
        return 1 / x**2
    
def g(x):
    if (x == 0):
        return None
    else:
        return 2 / x**2

def h(x):
    if (x == 0):
        return None
    else:
        return 3 + 2 / x**2
```

Next we plot the graphs of $f$, $g$ and $h$:

```python
import matplotlib.pyplot as plt
import numpy as np

fig, axes = plt.subplots(1,3)
fig.tight_layout(w_pad=5)
fig.figsize=[12, 8]
x = np.arange(-10, 10)
funcs = [f, g, h]
styles = ['red', 'blue', 'green']
labels = ['f', 'g', 'h']

for i in range(3):
    axes[i].plot(x, list(map(funcs[i], x)), c = styles[i])
    axes[i].grid(True)
    axes[i].set_xlabel('x')
    axes[i].set_ylabel('${}(x)$'.format(labels[i]))
    axes[i].set_xticks(np.arange(-10, 10, 5))
            
plt.show()
```

The figure below shows all three graphs on the same pair of axes.

```python
x1 = np.arange(-15, 15)
for i in range(3):
    plt.plot(x1, list(map(funcs[i], x1)), c = styles[i])
    

plt.legend(labels)
plt.grid(True) 
plt.show()
```
