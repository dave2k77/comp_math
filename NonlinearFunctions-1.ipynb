# Graph of Nonlinear Functions and their Transformations I

## Introduction

Nonlinear functions and graphs appear in numerous applications of mathematics, the mathematical sciences, engineering and the social sciences. It is therefore beneficial for scientists and mathematician to understand the theory of nonlinear functions and be versed in the mathematical operations and transformations associated with them. In this section we will take a short but in-depth look at certain types of nonlinear functions, their graph and simple transformations that they can undergo.

**Example 1**

Consider the function $f(x) = \frac{1}{x}$ where $x \ne 0$. This function is called a `reciprocal` function. It is in fact the simplest type. Let's look at its graph.

```python
# Define the function f(x) = 1/x
def f(x):
    if (x == 0):
        return None
    else:
        return 1/x
```

```python
# Plotting the graph of f(x)
import matplotlib.pyplot as plt
import numpy as np

x = np.arange(-25, 25)
plt.plot(x, list(map(f, x)))
plt.xlabel('x')
plt.ylabel('$f(x)$')
plt.title('Graph of $f(x) = {1}/{x}$')
plt.grid(True)
plt.show()
```

From the graph shown above you can easily deduce that $f(x)$ has a vertical asymptote at $x = 0$ and horizontal asymptote at $f(x) = 0$.

Now consider the function $g(x) = 1 + \frac{1}{x}$ where $x \ne 0$. We will compare both the graph of $f(x)$ and $g(x)$ below.

```python
def g(x):
    if (x == 0):
        return None
    else:
        return 1 + 1/x
```

```python
fig, axes = plt.subplots(1, 2)
fig.tight_layout(w_pad=5)
axes[0].plot(x, list(map(f, x)))
axes[0].set_xlabel("x")
axes[0].set_ylabel("$f(x)$")
axes[0].set_title("Graph of $f(x)$")
axes[0].grid(True)
axes[1].plot(x, list(map(g, x)))
axes[1].set_xlabel("x")
axes[1].set_ylabel("$g(x)$")
axes[1].set_title("Graph of $g(x)$")
axes[1].grid(True)
plt.show()
```

Notice that while the vertical asymptotes are the same, the horizontal asymptote for $g(x)$ is 1 compared to $f(x)$ which has horizontal asymptote being 0.

Making an algebraic comparison, we see that $g(x) = 1 + f(x)$. $g(x)$ represents an upward translation of the function $f(x)$ by 1 unit.


**Generalisation**

In general, if $F(x) = f(x) + A$ where $A > 0$, then $F(x)$ is an upward translation (along the y-axis) by $A$ units. If $A < 0$, then the translation is downward (along the y-axis) by $A$ units.

**Example 2**

Consider, now, the functions $h(x) = x^2$ and $k(x) = (x + 2)^2$. The figure below shows the graphical comparison between $h(x)$ and $k(x)$.

```python
def h(x):
    return x**2

def k(x):
    return (x + 2)**2
```

```python
x1 = np.arange(-5, 5)
fig, [ax1, ax2] = plt.subplots(1, 2)
fig.tight_layout(w_pad=5)
ax1.plot(x1, list(map(h, x1)))
ax1.set_xticks([i for i in range(-5, 5, 1)])
ax1.set_yticks([i for i in range(0, 25, 5)])
ax1.set_xlabel("x")
ax1.set_ylabel("$h(x)$")
ax1.set_title("Graph of $h(x)$")
ax1.grid(True)
ax2.plot(x1, list(map(k, x1)))
ax2.set_xticks([i for i in range(-5, 5, 1)])
ax2.set_yticks([i for i in range(0, 25, 5)])
ax2.set_xlabel("x")
ax2.set_ylabel("$k(x)$")
ax2.set_title("Graph of $k(x)$")
ax2.grid(True)
plt.show()
```

Notice that the function $k(x)$ is just a leftward translation of the function $h(x)$ by 2 units.

**Generalisation**

In general if a function $G(x) = g(x + A)$ where $A > 0$, then $G(x)$ is a leftward (along the x-axis) translation of the function $g(x)$ by $A$ units. If $A < 0$, then the translation is rightward (along the x-axis).
